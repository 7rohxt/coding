### Problem: Remove Duplicates from Sorted Array

#### Problem Statement
You are given a **sorted** array of integers `nums`. Remove the duplicate elements **in-place** such that each unique element appears only once. The **relative order** of the elements must be preserved.

After removing the duplicates, return the **new length** of the array. The first `k` elements of `nums` should contain the unique values, where `k` is the returned length.

You do not need to modify elements beyond the first `k` positions.

#### Input
- `nums`: a sorted list of integers

#### Output
- An integer `k` representing the number of unique elements after removal

#### Examples
```text
Input:  [1,1,2]
Output: 2
nums becomes [1,2,_]


### Initial Solution

````python
def remove_duplicates (nums):
    if len(nums) == 1:
        return 1

    pos = 1
    for i in range(1, len(nums)):
        if nums[i] != nums[i-1]:
            nums[pos] = nums[i]
            pos += 1
    

    k = len(nums[:pos])

    return k
print(remove_duplicates ([1,1,2,4,4,4,5,6,9]))
````

### Optimal

````python
def remove_duplicates(nums):
    if len(nums) == 0:
        return 0

    pos = 1
    for i in range(1, len(nums)):
        if nums[i] != nums[i-1]:
            nums[pos] = nums[i]
            pos += 1

    return pos
````
