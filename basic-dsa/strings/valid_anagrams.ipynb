### Problem: Valid Anagram

#### Problem Statement
Given two strings `s` and `t`, determine whether `t` is an **anagram** of `s`.

A string `t` is an anagram of `s` if it contains the **same characters** with the **same frequencies**, but possibly in a different order.

Return `True` if `t` is an anagram of `s`. Otherwise, return `False`.

---

#### Input
- `s`: a string
- `t`: a string

#### Output
- `True` if `t` is an anagram of `s`
- `False` otherwise

---

#### Example

```text
Input:  s = "anagram", t = "nagaram"
Output: True
Explanation: Both strings contain the same characters with the same counts.


### Method #1 - Create two hashmaps

````python
def valid_anagrams(s, t):
    s_map = {}
    t_map = {}

    for i in range(len(s)):
        s_map[s[i]] = s_map.get(s[i], 0) + 1

    for i in range(len(t)):
        t_map[t[i]] = t_map.get(t[i], 0) + 1

    return s_map == t_map

print(valid_anagrams("anagram","nagaram"))
````

### Method 2 - Subtract from a single hashmap

````python
def valid_anagrams(s, t):
    freq_map = {}

    for i in range(len(s)):
        freq_map[s[i]] = freq_map.get(s[i], 0) + 1
    
    for i in range(len(t)):
        if t[i] not in freq_map:
            return False
        
        freq_map[t[i]] -= 1

        if freq_map[t[i]] < 0:
            return False

    return True

print(valid_anagrams("anagram","nagaram"))
````
