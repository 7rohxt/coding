### Problem: First Non-Repeating Element

#### Problem Statement  
You are given an array of integers `nums`.

Return the **first element** in the array that appears **exactly once**.  
If no such element exists, return `-1`.

The order of elements in the array **must be preserved** when determining the first non-repeating element.

---

#### Input
- `nums`: a list of integers

#### Output
- An integer representing the first non-repeating element  
- Return `-1` if all elements repeat

---

#### Examples

```text
Input:  [9, 4, 9, 6, 7, 4]
Output: 6


### Initial solution

````python
def single_number (nums):
    unique_nums = []
    for num in nums:
    # for i, num in enumerate(nums):   -- this can also be used to get index and the value
        if num in unique_nums:                        # here num is the value and not the index. 
            unique_nums.remove(num)
        else:
            unique_nums.append(num)
    if len(unique_nums) == 0:
        return -1
    else:
        return unique_nums[0]

nums = [9, 4, 9, 6, 7, 4]
print(single_number(nums))

````

### Optimal Solution

````python
def single_number(nums):
    frequency = {}
    for num in nums:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1

    for num in nums:
        if frequency[num] == 1:
            return num
            
    return -1

nums = [9, 4, 9, 6, 7, 4]
print(single_number(nums))
        
````
