### Problem: Move All Zeros to the End

#### Problem Statement
You are given an array of integers `nums`. Move all `0`s to the end of the array while maintaining the relative order of the non-zero elements. The operation must be performed in-place without making a copy of the array.

#### Input
- `nums`: a list of integers

#### Output
- Modify the input array in-place so that all zeros are moved to the end

#### Examples
```text
Input:  [0,1,0,3,12]
Output: [1,3,12,0,0]


### My Solution

````python
def move_zeros(nums):
    non_zeros = []
    for num in nums:
        if num != 0:
            non_zeros.append(num)

    l_n = len(nums)
    l_nz = len(non_zeros)
    move_zeros = non_zeros

    add_zeros = l_n - l_nz

    for i in range (add_zeros):
        move_zeros.append(0)
    
    return move_zeros   
````

````python
nums = [0,1,0,3,12]
print(move_zeros(nums))
````
