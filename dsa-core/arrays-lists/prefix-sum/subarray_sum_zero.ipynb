### Problem: Check if a Subarray with Sum = 0 Exists

#### Problem Statement
You are given an array of integers `nums` (which may contain positive, negative, and zero values).

Determine whether there exists a **non-empty contiguous subarray** whose sum is equal to `0`.

Return `True` if such a subarray exists, otherwise return `False`.

---

#### Input
- `nums`: a list of integers

#### Output
- `True` if there exists at least one non-empty subarray with sum equal to `0`
- `False` otherwise

---

#### Example

```text
Input:  [4, 2, -3, 1, 6]
Output: True
Explanation: The subarray [2, -3, 1] has sum 0.


### Brute force

````python
def cur_sum_zero_subarray(nums):

    for i in range (len(nums)):
        cur_sum = 0 
        for j in range(i, len(nums)):
            cur_sum += nums[j]
            if cur_sum == 0:
                return True
    return False

print(cur_sum_zero_subarray([4,2,-3,1,6]))
````

### optimal

````python
def cur_sum_zero_subarray(nums):
    prefix_sum = 0
    prefix_sum_seen = set()

    for i in range(len(nums)):
        prefix_sum += nums[i]
    
        if prefix_sum == 0:
            return True
        
        if prefix_sum in prefix_sum_seen:
            print(prefix_sum_seen)
            return True

        else:
            prefix_sum_seen.add(prefix_sum)

    return False
        
print(cur_sum_zero_subarray([4,2,-3,1,6]))
````
