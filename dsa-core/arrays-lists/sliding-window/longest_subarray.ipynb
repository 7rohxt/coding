### Problem: Longest Subarray with Sum ≤ k

#### Problem Statement
You are given an array of **positive integers** `nums` and an integer `k`.  

Find the **length of the longest contiguous subarray** such that the sum of its elements is **less than or equal to `k`**.

Return the maximum length of such a subarray.

---

#### Input
- `nums`: a list of **positive integers**
- `k`: an integer representing the maximum allowed sum

#### Output
- An integer representing the **maximum length** of a contiguous subarray whose sum is ≤ `k`

---

#### Constraints
- `1 ≤ len(nums) ≤ 10^5`
- `1 ≤ nums[i] ≤ 10^4`
- `1 ≤ k ≤ 10^9`

---

#### Example

```text
Input:  nums = [1,2,3,4,5], k = 7
Output: 3

Explanation:
The longest subarray with sum ≤ 7 is [1,2,3] (sum = 6).


### Initial Solution

````python
# Wrong solution, It doesn't check the longest sequence. It only checks the starting sequence

def longest_subarray(nums, k):
    sum = 0 
    for i in range(len(nums)):
        sum += nums[i]
        if sum > k:
            return i

print(longest_subarray([1,2,3,4,5], 7))
        
````

### Sliding window - Initial solution

````python
def longest_subarray(nums, k):
    left = 0 #0     #0
    cur_sum = 0 #1  #3
    max_len = 0 #1  

    for right in range (len(nums)):
        cur_sum += nums[right]
        

        while cur_sum > k:
            cur_sum -= nums[left]
            left +=1 
    
        max_len = max(max_len, right - left + 1)
    return max_len

print(longest_subarray([1,2,3,4,5], 7))
        
````
