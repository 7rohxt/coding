#### 3. Sum of first n natural numbers

```python
# While loop --------- time - o(n), space - o(1)
n = 10
i = 1
sum = 0
while i <= n:
    sum += i
    i +=1
print(sum)

# Using Reccursion----------- time - o(n), space - o(n)
def n_sum(n):
    if n == 1:
        return 1
    return n + n_sum(n-1)
print(n_sum(10))

# Using formula --------------- time - o(1), space - o(1)
print(int((n*(n+1))/2))
```

#### 4. Sum of n given a range included

#### 6. Print Greatest of three nums

```python
num1, num2, num3 = 1, 6, 2

# Using ternary
max = num1 if num1>=num2 else num2
max = num3 if num3>=max else max
print(max)
```

#### 7. Leap year

```python
year = 2200

# using Ternary
print(True if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0) else False)

# Using Lambda function
isLeap = lambda year: True if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0) else False 
print(isLeap(year))
```

#### 8. Check Prime

```python
num = 15
flag = 0
if num < 2:
    flag = 1
elif num == 2:
    flag = 0
elif num % 2 == 0:
    flag = 1
else :
    # for i in range(3, int(num**0.5) + 1, 2): --> more efficient
    for i in range(3, (num+1)//2, 2):
        if num % i == 0:
            flag = 1
            break
print(True if flag == 0 else False)
```

#### 9. Print Prime Numbers in a Given Range

```python
low, high = 2, 10

def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, (n//2) + 1 , 2 ):
        if n % i == 0:
            return False
    return True

for i in range (low, high+1):
    if is_prime(i):
        print(i)
```
