#### 1. Postivie or Negative 

```python
# Using Ternary --------------- time - o(1), space - o(1)
num = 15
print("Positive" if num >= 0 else "Negative") 
```

#### 2. Even or Odd 

```python
# Using Ternary --------------- time - o(1), space - o(1)
num = 15
print("Even" if num%2 == 0 else "Odd")
```

#### 3. Sum of first n natural numbers

```python
# While loop --------- time - o(n), space - o(1)
n = 10
i = 1
sum = 0
while i <= n:
    sum += i
    i +=1
print(sum)

# Using Reccursion----------- time - o(n), space - o(n)
def n_sum(n):
    if n == 1:
        return 1
    return n + n_sum(n-1)
print(n_sum(10))

# Using formula --------------- time - o(1), space - o(1)
print(int((n*(n+1))/2))
```

#### 4. Sum of n given a range included

```python
# Using For loop 
num1, num2 = 3, 6
sum = 0
for i in range(num1, num2+1):
    sum += i
print(sum)

# Using Formula
print(int((num2*(num2+1))/2) - int((num1*(num1+1))/2) + num1)
```

#### 5. Print greatest of 2 nums

```python
# Using If else
num1, num2 = 3, 6
if num1 > num2:
    print(num1)
else:
    print(num2)

# Using ternary
print(num1 if num1>num2 else num2)
```
