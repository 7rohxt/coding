```python
"""
QUESTION: Golden House Manager Puzzle

A house manager is inspecting a row of N rooms. Each room contains a certain
number of gold coins. The number of coins in each room is given in an array
C of size N.

The manager can start from any room i and move only to the next room
(i → i+1 → i+2 ...), forming a continuous sequence of rooms.

Your task is to find the starting and ending room indices such that the
sum of coins in that continuous segment is exactly equal to K.

If multiple such segments exist, return the one with the smallest
starting index.

If no such segment exists, print -1.

Input Format:
The first line contains two integers N and K.
The second line contains N space-separated integers representing the
number of coins in each room.

Output Format:
Print two integers representing the starting index and ending index
(1-based) of the continuous rooms whose coin sum equals K.
If no valid segment exists, print -1.

Constraints:
1 ≤ N ≤ 10^5
0 ≤ C[i] ≤ 10^9
1 ≤ K ≤ 10^9

Example 1:

Input:
5 12
2 3 7 5 1

Output:
2 4

Explanation:
Rooms [2,3,4] contain coins [3,7,5].
Their sum = 3 + 7 + 5 = 15 → not valid.

Rooms [2,3] contain coins [3,7].
Their sum = 10 → not valid.

Rooms [2,3,4,5] contain coins [3,7,5,1].
Their sum = 16 → not valid.

Rooms [2,3,4] → Actually the valid segment is [3,4]:
7 + 5 = 12.

Thus start = 3, end = 4.

Example 2:

Input:
6 9
1 2 3 4 5 6

Output:
2 4

Explanation:
Rooms [2,3,4] contain coins [2,3,4].
Their sum = 9.
"""
```

```python
def findRooms(N, K, coins):
    left = 0
    cur_sum = 0

    for right in range(N):
        cur_sum += coins[right]

        while cur_sum > K:
            cur_sum -= coins[left]
            left +=1
        
        if cur_sum == K:
            return ([left+1, right+1])
    return -1

N, K = map(int, input().split())
coins = list(map(int,input().split()))

ans = findRooms(N, K, coins)
if ans == -1:
    print(-1)
else:
    print(*ans)
```

```python
"""
QUESTION: Horse Betting Strategy

Bob is attending a horse race where there are N horses lined up in a row.
Each horse has a betting price associated with it. The prices are given
in an array P of size N.

Bob wants to place bets on a continuous sequence of horses. However,
he has a rule: the total amount he bets must be strictly less than K.

Your task is to determine the maximum number of horses Bob can bet on
in a continuous sequence such that the total betting price of those
horses is strictly less than K.

Input Format:
The first line contains two integers N and K.
The second line contains N space-separated integers representing the
betting prices of the horses.

Output Format:
Print a single integer representing the maximum length of a continuous
sequence of horses whose total betting price is strictly less than K.

Constraints:
1 ≤ N ≤ 10^5
1 ≤ P[i] ≤ 10^9
1 ≤ K ≤ 10^9

Example:

Input:
6 10
2 3 5 4 2 1

Output:
3

Explanation:
Possible sequences with sum < 10:

[2,3] → sum = 5 (length 2)
[2,3,5] → sum = 10 (not allowed, must be < 10)
[3,5] → sum = 8 (length 2)
[3,5,4] → sum = 12 (not allowed)
[5,4] → sum = 9 (length 2)
[4,2,1] → sum = 7 (length 3)

The maximum length of a valid continuous sequence is 3.
"""
```

```python
def maxHourses(N, K, prices):
    max_len = 0
    cur_len = 0
    cur_price = 0
    left = 0

    for right in range(N):
        cur_price += prices[right]

        while cur_price >= K:
            cur_price -= prices[left]
            left += 1
        
        cur_len = right - left + 1
        max_len = max(max_len, cur_len)
    
    return max_len
        

N, K = map(int, input().split())
prices = list(map(int, input().split()))

print(maxHourses(N, K, prices))
```
