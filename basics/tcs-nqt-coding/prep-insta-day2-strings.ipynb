## String EASY (Basic Traversal & Built-in Usage)

1. Check if a given string is palindrome | Input: "madam"  
2. Count number of vowels, consonants, spaces | Input: "hello world"  
3. Find ASCII value of a character | Input: 'A'  
4. Remove all vowels from string | Input: "education"  
5. Remove spaces from string | Input: "hello world"  
6. Remove characters except alphabets | Input: "he11o@w0rld!"  
7. Reverse a string | Input: "python"  
8. Change case of each character | Input: "PyThOn"  
9. Concatenate one string to another | Input: str1="hello", str2="world"  
10. Count number of words in a string | Input: "welcome to python programming"  
11. Sort characters in a string | Input: "python"

```python
s = "madam"
left = 0
right = len(s) - 1
while left < right:
    if s[left] == s[right]:
        left += 1
        right -= 1
    else:
        print(False)
        break
else:
    print(True)
```

```python
# 2. Count number of vowels, consonants, spaces | Input: "hello world"  

s = "hello world"
vow_count, conso_count, space_count = 0, 0, 0
vowels_set = {"a", "e", "i", "o", "u"}  # ---  set

for ch in s:
    if ch.lower() in vowels_set:
        vow_count += 1
    elif ch == " ":
        space_count += 1
    elif ch.isalpha():
        conso_count += 1
print(vow_count, conso_count, space_count)
```

```python
# 3. Find ASCII value of a character | Input: 'A'  
ch = 'A'

ascii_val = ord(ch)
print(ascii_val)

ascii_val2 = 76
print(chr(ascii_val2))
```

```python
# 4. Remove all vowels from string | Input: "education"  

s = "education"  
res = []
vowels = {"a", "e", "i", "o", "u"}
for ch in s:
    if ch.lower() not in vowels:
        res.append(ch)
print("".join(res))
```

```python
# 5. Remove spaces from string | Input: "hello world"  

s = "hello world"

print("".join(ch for ch in s if ch != " "))
print(s.replace(" ", ""))
```

```python
# 6. Remove characters except alphabets | Input: "he11o@w0rld!"  
s = "he11o@w0rld!"  

print("".join(ch for ch in s if ch.isalpha()))
```

```python
# 7. Reverse a string | Input: "python" 
s = "python"
s_list = list(s)
l, r = 0, len(s_list) - 1
while l < r:
    s_list[l], s_list[r] = s_list[r], s_list[l]
    l += 1
    r -= 1
print("".join(s_list))
```

```python
# 8. Change case of each character | Input: "PyThOn"  

s = "PyThOn" 
res = list(s)

for i in range(len(res)):
    if res[i].islower():
        res[i] = res[i].upper()
    elif res[i].isupper():
        res[i] = res[i].lower()
print("".join(res))
```

```python
# 9. Concatenate one string to another | Input: str1="hello", str2="world"  

str1="hello"
str2="world" 

res = list(str1)

for ch in str2:
    res.append(ch)
print("".join(res))
```

```python
# 10. Count number of words in a string | Input: "welcome to python programming"  

s = "welcome to python programming"  

# Solution 1
print(len(s.split()))

# Solution 2 (without split)
count = 0
for ch in s:
    if not ch:
        print("0")
        break
    if ch == " ":
        count += 1
print(count + 1)
```

```python
# 11. Sort characters in a string | Input: "python"

s = "python"
print("".join(sorted(s)))
```

## String MEDIUM (Hashing, Frequency, Patterns)

1. Calculate frequency of characters | Input: "programming"  
2. Find non-repeating characters | Input: "swiss"  
3. Return maximum occurring character | Input: "banana"  
4. Remove all duplicates from string | Input: "programming"  
5. Print all duplicates in string | Input: "programming"  
6. Remove characters from first string present in second | Input: str1="computer", str2="cat"  
7. Find largest word in string | Input: "python is a powerful language"  
8. Reverse words in a string | Input: "hello world python"  
9. Sum of numbers in a string | Input: "1abc23"  
10. Find substring and display starting position | Input: string="hello world", substring="world"  
11. Change every letter with next lexicographic alphabet | Input: "abcd"  
12. Find word with highest repeated letters | Input: "apple banana mango"

```python
# 1. Calculate frequency of characters | Input: "programming"  

s = "programming"
from collections import Counter
freq = dict(Counter(s))
print(freq)

freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)

freq = {}
for ch in s:
    if not ch in freq:
        freq[ch] = 0
    freq[ch] += 1
print(freq)
```

```python
# 2. Find non-repeating characters | Input: "swiss"  

s = "swiss"
freq = Counter(s)
res = []
for ch in s:
    if freq[ch] == 1:
        res.append(ch)
print(res)
```

```python
# 3. Return maximum occurring character | Input: "banana"  

s = "banana"
freq = Counter(s)

max_count = max(freq.values())
for ch in s:
    if freq[ch] == max_count:
        print(ch)
        break

max_char = max(freq, key=freq.get) # --------- remember this
print(max_char)
```

```python
# 4. Remove all duplicates from string | Input: "programming"  

s = "programming"
res = []
seen = set()

for ch in s:
    if not ch in seen:
        res.append(ch)
        seen.add(ch)
print("".join(res))
```

```python
# 5. Print all duplicates in string | Input: "programming"  

s = "programming"
res = []
freq = Counter(s)

for ch in s:
    if freq[ch] > 1:
        res.append(ch)

print("".join(res))
```

```python
# 6. Remove characters from first string present in second | Input: str1="computer", str2="cat"  

str1="computer"
str2="cat"  

res = []
unique_str2 = set(str2)

for ch in str1:
    if not ch in unique_str2:
        res.append(ch)
print("".join(res))
```

```python
# 7. Find largest word in string | Input: "python is a powerful language"  

s = "python is a powerful language" 
s1 = s.split()
largest_word = ""
for word in s1:
    if len(word) > len(largest_word):
        largest_word = word
print(largest_word)

s2 = s.split()
largest_word = max(s2, key=len)
print(largest_word)
```

```python
# 8. Reverse words in a string | Input: "hello world python"  

s = "hello world python"  

# Sol 1
s_list = s.split()
l, r = 0, len(s_list) -1
while l < r:
    s_list[l], s_list[r] = s_list[r], s_list[l]
    l += 1
    r -= 1
print(" ".join(s_list))

# Sol 2 using [::-1]
reversed_list = s.split()[::-1]
print(" ".join(reversed_list))
```

```python
# 9. Sum of numbers in a string | Input: "1abc23"
  
s = "1abc23"
total = 0
for ch in s:
    if ch.isnumeric():
        total += int(ch)
print(total)
```

```python
# 10. Find substring and display starting position | Input: string="hello world", substring="world"  

string="hello world"
substring="world"  

for i in range(len(string) - len(substring) + 1):
    if string[i : i+len(substring)] == substring:
        print(i)
        break
```

```python
# 11. Change every letter with next lexicographic alphabet | Input: "abcd"  
s = "abcdz"
res = []
for ch in s:
    res.append(chr(ord(ch)+1))
print("".join(res))

# if needed to handle z -> a
for i in range(len(res)):
    if res[i] == "{":
        res[i] = "a"
print("".join(res))
```

```python
# 12. Find word with highest repeated letters | Input: "apple banana mango"
from collections import Counter

s = "apple banana mango"
s = s.split()

result = ""
max_duplicates = -1

for word in s:
    freq = Counter(word)
    cur_duplicates = max(freq.values())

    if cur_duplicates > max_duplicates:
        max_duplicates = cur_duplicates
        result = word

print(result)
```
