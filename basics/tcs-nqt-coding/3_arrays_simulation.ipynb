```python
"""
QUESTION: Verbal Kho-Kho

Alice gives a digit to the first friend. Each friend communicates the digit
to the next friend without speaking.

Finally the last friend writes the digit. If the digit differs from the
original, find how many friends misunderstood the digit.

Input
N
array D representing digits understood by each friend

Output
number of friends who misunderstood.
"""
```

```python
def verbalKhokho (N, D):
    count = 0
    for i in range(1, N):
        if D[i] != D[i - 1]:
            count += 1
    return count
N =  int(input().strip())
D = list(map(int, input().split()))

print(N)
print(D)
print(verbalKhokho(N, D))
```

```python
"""
QUESTION: Maximum Guests on Cruise

A cruise ship is hosting a party that lasts for T hours. During each hour,
some guests enter the party and some guests leave.

You are given two arrays:

E[i] – the number of guests entering the party during hour i  
L[i] – the number of guests leaving the party during hour i

Assume that:
1. All guests who leave during hour i were already present at the party.
2. The party initially starts with 0 guests.
3. Guests who enter and leave during the same hour are counted in that hour.

Your task is to determine the maximum number of guests present at the party
at any point during the T hours.

Input:
T                      → number of hours the party lasts
E[0..T-1]              → array representing guests entering each hour
L[0..T-1]              → array representing guests leaving each hour

Output:
Return a single integer representing the maximum number of guests present
on the cruise at any time.

Example:

Input:
T = 5
E = [7, 0, 5, 1, 3]
L = [1, 2, 1, 3, 4]

Maximum guests present at any time = 8

Output:
8
"""
```

```python
def maxGuests(T, E, L):
    max_guests = 0
    cur_guests = 0
    for i in range(T):
        cur_guests = cur_guests + E[i] - L[i]
        max_guests = max(cur_guests, max_guests)
    return max_guests

T = int(input().strip())
E = list(map(int, input().split()))
L = list(map(int, input().split()))

print(maxGuests(T, E, L))
```

```python
"""
QUESTION: Balloon Color With Odd Frequency

Given N balloons, each with a color.

Find the first color that appears an odd number of times.
If all colors appear an even number of times, print "All are even".

Input:
N
N space-separated balloon colors

Output:
First color with odd frequency
or "All are even"

Example:

Input:
7
red blue red green blue green red

Output:
red
"""
```

```python
from collections import Counter

def firstColour(N, colours):
    colour_freq = Counter(colours)

    for colour in colours:
        if colour_freq[colour] % 2 != 0:
            return colour
    
    return "All are even"

N = int(input().strip())
colours = list(input().split())
print(firstColour(N, colours))
```

```python
"""
QUESTION: Counting Rock Samples

John is a geologist who needs to count rock samples before sending them
to a chemical laboratory. The laboratory accepts rock samples only within
certain size ranges measured in ppm (parts per million).

Given an array samples[] representing the sizes of rock samples and a
2D array ranges[][] representing acceptable ranges, determine how many
samples fall within each range.

For every range ranges[i][0] to ranges[i][1], count the number of rock
samples whose size lies within the range (inclusive).

Input Format:
The first line contains space-separated integers representing the
rock sample sizes.

The second line contains an integer R representing the number of ranges.

The next R lines contain two space-separated integers representing
the lower and upper bound of each range.

Output Format:
Print the count of samples falling within each range.

Example 1:

Input:
345 604 321 433 704 470 808 718 517 811
2
300 380
400 700

Output:
2 4

Explanation:
Range [300, 380] → samples: 345, 321 → count = 2
Range [400, 700] → samples: 433, 604, 517, 470 → count = 4


Example 2:

Input:
400 567 890 765 987
2
300 380
800 1000

Output:
0 2
"""
```

```python
def countRocks(samples, R, ranges):
    res = []
    for r in ranges:
        cur_count = 0
        low, high = r[0], r[1]
        for sample in samples:
            if sample >= low and sample <= high:
                cur_count += 1
        res.append(cur_count)
    return res

samples = list(map(int, input().split()))
R = int(input().strip())
ranges = []
for _ in range(R):
    low, high = map(int, input().split())
    ranges.append([low, high])

print(*countRocks(samples, R, ranges))
```

```python
"""
QUESTION: Cheating Detection in Exam

In a class there are N students seated in a row. The marks obtained by
each student are stored in an array M of size N.

The examination board suspects cheating if two adjacent students have
exactly the same marks. In such a case, the latter student (the student
to the right) is removed from the merit list.

You are given K re-evaluation operations. In each operation, the mark of
a student may change.

After every update, determine how many students remain in the merit list
after removing all students who violate the cheating rule.

Note:
If M[i] == M[i-1], then student i is removed from the merit list.

Input Format:
The first line contains two integers N and K.
The second line contains N space-separated integers representing the
marks of the students.
The next K lines each contain two integers X and Y:
    X → index of the student (1-based)
    Y → new mark after re-evaluation.

Output Format:
After each update, print the number of students remaining in the merit list.

Constraints:
1 ≤ N, K ≤ 10^5
0 ≤ M[i], Y ≤ 10^9

Example:

Input:
5 2
10 20 20 30 40
2 10
3 30

Output:
4
3

Explanation:

Initial Marks:
[10, 20, 20, 30, 40]

Students 2 and 3 have equal marks.
Student 3 (the latter one) is removed.

Remaining students = 5 − 1 = 4

Update 1:
Student 2 → mark becomes 10

Marks become:
[10, 10, 20, 30, 40]

Students 1 and 2 have equal marks.
Student 2 is removed.

Remaining students = 4

Update 2:
Student 3 → mark becomes 30

Marks become:
[10, 10, 30, 30, 40]

Students 1 and 2 have equal marks → remove student 2
Students 3 and 4 have equal marks → remove student 4

Remaining students = 5 − 2 = 3
"""
```

```python
def cheating(N, K, marks, updates):
    res = []

    for update in updates:
        marks[update[0]-1] = update[1]
        cur_cheat_count = N
        for i in range(1, N):
            if marks[i] == marks[i-1]:
                cur_cheat_count -= 1
        res.append(cur_cheat_count)
    
    return res

N, K = map(int, input().split())
marks = list(map(int, input().split()))

updates = []
for _ in range(K):
    i, mark = map(int, input().split())
    updates.append([i, mark])

print(*cheating(N, K, marks, updates))
```
