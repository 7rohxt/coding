```python
"""
QUESTION: Verbal Kho-Kho

Alice gives a digit to the first friend. Each friend communicates the digit
to the next friend without speaking.

Finally the last friend writes the digit. If the digit differs from the
original, find how many friends misunderstood the digit.

Input
N
array D representing digits understood by each friend

Output
number of friends who misunderstood.
"""
```

```python
def verbalKhokho (N, D):
    count = 0
    for i in range(1, N):
        if D[i] != D[i - 1]:
            count += 1
    return count
N =  int(input().strip())
D = list(map(int, input().split()))

print(N)
print(D)
print(verbalKhokho(N, D))
```

```python
"""
QUESTION: Maximum Guests on Cruise

A cruise ship is hosting a party that lasts for T hours. During each hour,
some guests enter the party and some guests leave.

You are given two arrays:

E[i] – the number of guests entering the party during hour i  
L[i] – the number of guests leaving the party during hour i

Assume that:
1. All guests who leave during hour i were already present at the party.
2. The party initially starts with 0 guests.
3. Guests who enter and leave during the same hour are counted in that hour.

Your task is to determine the maximum number of guests present at the party
at any point during the T hours.

Input:
T                      → number of hours the party lasts
E[0..T-1]              → array representing guests entering each hour
L[0..T-1]              → array representing guests leaving each hour

Output:
Return a single integer representing the maximum number of guests present
on the cruise at any time.

Example:

Input:
T = 5
E = [7, 0, 5, 1, 3]
L = [1, 2, 1, 3, 4]

Explanation:
Hour 1: Guests = 0 + 7 - 1 = 6
Hour 2: Guests = 6 + 0 - 2 = 4
Hour 3: Guests = 4 + 5 - 1 = 8
Hour 4: Guests = 8 + 1 - 3 = 6
Hour 5: Guests = 6 + 3 - 4 = 5

Maximum guests present at any time = 8

Output:
8
"""
```

```python
def maxGuests(T, E, L):
    max_guests = 0
    cur_guests = 0
    for i in range(T):
        cur_guests = cur_guests + E[i] - L[i]
        max_guests = max(cur_guests, max_guests)
    return max_guests

T = int(input().strip())
E = list(map(int, input().split()))
L = list(map(int, input().split()))

print(maxGuests(T, E, L))
```
