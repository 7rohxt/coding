## Day 5 – TCS Practice Questions

### 1. Handshake Count

Before the outbreak of COVID-19, a meeting was held in a room where **N people** attended.  
Everyone greeted each other by **shaking hands exactly once**.

Your task is to determine the **total number of handshakes** that happened in the meeting.

Each handshake occurs between **two different people**, and no pair shakes hands more than once.

#### Input Format
- The first line contains an integer **T**, the number of test cases.
- The next **T lines** each contain an integer **N**, the number of people in the meeting.

#### Output Format
- For each test case, print the **total number of handshakes**.

### 2. Cheating Detection in Exam

John is a mathematics teacher. His class has **N students**, seated one behind another based on their **roll numbers (1 to N)**.

The marks of the students are stored in an array **M[ ]**.

If **two adjacent students have the same marks**, it is considered **cheating**, and the **student sitting behind (the latter student)** will be removed from the merit list.

Later, the marks of **K students** may change due to **re-evaluation**.

Each re-evaluation is given as:


X Y


- **X** → index of the student
- **Y** → new mark for that student

After each re-evaluation:
- update the student's marks
- remove any student who is detected cheating (adjacent equal marks)

Your task is to determine the **maximum number of students remaining in the merit list after each re-evaluation**.


### 3. Parking Lot Occupancy

A shopping mall has a **parking lot represented as a matrix**.

The parking lot has **R rows and C columns**.

Each parking space is represented by:
- **0 → empty parking space**
- **1 → occupied parking space**

Your task is to determine the **index of the row that contains the maximum number of occupied parking spaces (1s)**.

If multiple rows have the same maximum number of occupied spaces, return the **first such row index**.

#### Input
- A matrix of size **R × C** containing only **0s and 1s**

#### Output
- The **row index (R)** with the maximum number of occupied parking spaces.
