## Array EASY (Basic Traversal & Simple Logic)

1. Find the smallest number in an array | Input: [4, 2, 7, 1, 9]  
2. Average of all elements in an array | Input: [2, 4, 6, 8]  
3. Search an element in an array (Linear Search) | Input: arr = [5, 3, 8, 2, 9], target = 8  
4. Reverse a given array | Input: [1, 2, 3, 4, 5]  
5. Adding element in an array | Input: arr = [1, 2, 3, 4], element = 5  
6. Count frequency of each element in an array | Input: [1, 2, 2, 3, 3, 3]  
7. Find all repeating elements in an array | Input: [1, 2, 3, 2, 4, 1, 5]  
8. Find all non-repeating elements in an array | Input: [1, 2, 2, 3, 4, 4, 5]  
9. Remove duplicates from a sorted array | Input: [1, 1, 2, 2, 3, 4, 4, 5]

```python
# 1. Smallest number in an array

def smallestNum(nums):
    smallest_num = nums[0]
    for num in nums:
        if num < smallest_num:
            smallest_num = num
    return smallest_num

print(smallestNum([4,2,7,1,9]))

print(min([4,2,7,1,9]))
    
```

```python
# 2. Average of all elements in an array | Input: [2, 4, 6, 8]  

def avgElements(nums):
    total = 0
    for num in nums:
        total += num
    avg = total//len(nums)
    return avg

print(avgElements([4,2,7,1,11]))
```

```python
# 3. Search an element in an array (Linear Search) | Input: arr = [5, 3, 8, 2, 9], target = 8  

arr = [5, 3, 8, 2, 9]
target = 8 
for i, num in enumerate(arr):
    if num == target:
        print(i)
```

```python
# 4. Reverse a given array | Input: [1, 2, 3, 4, 5]  

nums = [1, 2, 3, 4, 5]
n = len(nums)
l, r = 0, n-1
while l < r:
    nums[l], nums[r] = nums[r], nums[l]
    l += 1
    r -=1
print(nums)
```

```python
# 5. Adding element in an array | Input: arr = [1, 2, 3, 4], element = 5  

nums = [1, 2, 3, 4]
element = 5
nums.append(element)
print(nums)
```

```python
# 6. Count frequency of each element in an array | Input: [1, 2, 2, 3, 3, 3]  

nums = [1, 2, 2, 3, 3, 3]  
freq = {}
for num in nums:
    if not num in freq:
        freq[num] = 0
    freq[num] += 1
print(freq)

from collections import Counter
freq = dict(Counter(nums))
print(freq)
```

```python
# 7. Find all repeating elements in an array | Input: [1, 2, 3, 2, 4, 1, 5]  

nums = [1, 2, 3, 2, 4, 1, 5] 
repeating_nums = []
seen = set()
for num in nums:
    if num not in seen:
        seen.add(num)
    else:
        repeating_nums.append(num)
print(repeating_nums)    
```

```python
# 8. Find all non-repeating elements in an array | Input: [1, 2, 2, 3, 4, 4, 5]  

nums = [1, 2, 3, 2, 4, 1, 5] 
non_repeating_nums = []

from collections import Counter
freq = dict(Counter(nums))

for num in nums:
    if freq[num] == 1:
        non_repeating_nums.append(num)
print(non_repeating_nums)
```

```python
# 9. Remove duplicates from a sorted array | Input: [1, 1, 2, 2, 3, 4, 4, 5]

nums = [1, 1, 2, 2, 3, 4, 4,4,4, 4, 5]
res = [nums[0]]
for i in range(1, len(nums)):
    if nums[i] != nums[i-1]:   # ---> remember to handle edge cases
        res.append(nums[i])
print(res)
```

```python
# 1. Second smallest and second largest element | Input: [4, 2, 7, 1, 9]

nums = [4, 2, 7, 1, 9]
nums.sort()
print([nums[1], nums[-2]])

nums = [4, 2, 7, 1, 9]
smallest = second_smallest = float('inf')
largest = second_largest = float('-inf')

for i in range(len(nums)):
    if nums[i] < smallest:
        second_smallest = smallest
        smallest = nums[i]
    elif nums[i] < second_smallest:
        second_smallest = nums[i]

    if nums[i] > largest:
        second_largest = largest
        largest = nums[i]
    elif nums[i] > second_largest:
        second_largest = nums [i]
    
print(second_smallest, second_largest)
```

```python
# 2. Remove duplicates from unsorted array | Input: [3, 1, 2, 3, 4, 2, 5]

nums = [3, 1, 2, 3, 4, 2, 5]
res = []
seen = set ()
for num in nums:
    if not num in seen:
        seen.add(num)
        res.append(num)
print(res)

print(list(set(nums))) # --> if order is not an issue
```

```python
# 3. Find the median of the given array | Input: [7, 1, 3, 4, 5]

nums = [7, 1, 3, 4, 5]
nums.sort()
median = nums[len(nums) // 2]
print(median)
```

```python
# 4. Rearrange array in increasing-decreasing order | Input: [5, 2, 8, 1, 4, 7]

nums = [5, 2, 8, 1, 4, 7]
n = len(nums)
nums.sort()

middle = n // 2

l = middle
r = n - 1

while l < r:
    nums[l], nums[r] = nums[r], nums[l]
    l += 1
    r -= 1
print(nums)


nums = [5, 2, 8, 1, 4, 7]
nums.sort()
n = len(nums)
middle = len(nums) // 2
res = nums[:middle] + nums[n:middle-1:-1] # --> start:end:step (1 is step, '-' is direction)
res
```

```python
# 5. Sorting elements of an array by frequency | Input: [4, 4, 1, 2, 2, 2, 3]

nums = [4, 4, 1, 2, 2, 2, 3]

from collections import Counter

freq = Counter(nums)
nums.sort(key = lambda x: (freq[x], x))
nums
```

```python
# 6. Rotation of elements of array – left and right | Input: arr = [1, 2, 3, 4, 5], k = 1

# Left rotation

nums, k = [1, 2, 3, 4, 5], 2
# res = []
# for i in range(k, len(nums)):
#     res.append(nums[i])
# for i in range(k):
#     res.append(nums[i])
# print("left: ", res)

simple_left_res = nums[k:] + nums[:k]  # --- prefer this
print("simple left: ", simple_left_res)


# Right rotation

nums, k = [1, 2, 3, 4, 5], 2
# res = []
# for i in range(len(nums) - k, len(nums)):
#     res.append(nums[i])
# for i in range(len(nums)-k):
#     res.append(nums[i])
# print("right: ", res)

simple_right_res = nums[-k:] + nums[:-k]   # --> prefer this
print("simple right: ", simple_right_res)
```

```python
# 7. Finding equilibrium index of an array | Input: [1, 3, 5, 2, 2]
nums = [1, 3, 5, 2, 2]
left_sum = 0
total_sum = sum(nums)
for i in range(len(nums)):
    right_sum = total_sum - nums[i] - left_sum

    if left_sum == right_sum:
        print(i)
    
    left_sum += nums[i]
```

```python
# 8. Check if array is a subset of another array or not | Input: arr1 = [1, 2, 3, 4, 5], arr2 = [2, 4]
arr1 = [1, 2, 3, 4, 5]
arr2 = [2, 4]
left = 0

arr1_freq = Counter(arr1)
flag = 0
for num in arr2:
    if not num in arr1_freq:
        flag = 1
        break
    arr1_freq[num] -= 1

    if arr1_freq[num] < 0:
        flag = 1
        break
print(flag == 0)
```

```python
# 9. Find all symmetric pairs in array | Input: [(1,2), (3,4), (2,1), (5,6)]
pairs_list = [(1,2), (3,4), (2,1), (5,6), (6,5)]

seen = set ()
res = []

for a, b in pairs_list:
    if (b, a) in seen:
        res.append([(a,b), (b,a)])
    else:
        seen.add((a, b))
print(res)
```

```python
# 10. Replace each element of the array by its rank | Input: [40, 10, 20, 30]

nums = [40, 10, 20, 30]

# nums_sorted = nums[:]  ---  you dont need .sort() sorted better 1 line
# nums_sorted.sort() 

nums_sorted = sorted(set(nums)) # better coz rank based on unique (NEED TO USE SET ELSE 20, 20 WILL GIVE RANK 1,2)

rank_map = {}
for i in range(len(nums_sorted)):
    rank_map[nums_sorted[i]] = i + 1

for i in range(len(nums)):
    nums[i] = rank_map[nums[i]]

print(nums)
```
