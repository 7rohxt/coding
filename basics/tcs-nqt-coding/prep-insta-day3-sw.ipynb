## TCS Previous Year Coding Questions (Sliding Window / Subarray)

### Question 1: Golden House Manager Puzzle

This vacation you visited a golden house that has **N rooms** arranged in a sequence from **1 to N**.

Each room contains some **gold coins**. The owner wants to hire an intelligent manager, so he created a puzzle.

You must choose:

- one room to **enter**
- one room to **exit**

From any room you can either:

- **exit immediately**, or
- **move to the next room**

While visiting rooms you **must collect all coins from every room you enter**. You cannot skip collecting coins.

The owner wants you to collect **exactly K coins** when you exit.

If there are multiple solutions, you must return the one with the **smallest starting room number**.

Room numbering starts from **1 to N**.

**Task**

Find the **starting room index and ending room index** such that the **sum of coins in the continuous rooms equals K**.

**Hint**

Find a **continuous subsequence (subarray)** whose sum is exactly **K**.

---

Example

Input
Input

N = 5
coins = [1, 2, 3, 4, 5]
K = 9


Output

2 4


Explanation


2 + 3 + 4 = 9

### Question 2: Horse Betting Strategy

Bob is going to bet today on horse racing. There are **N horses** arranged in a sequence from **1 to N**.

The cost of betting on each horse is different. Bob can bet on any number of horses, but he believes that betting on a **continuous sequence of horses** gives him a better chance of winning.

If any horse on which Bob bets wins, he receives **K units of money**.

Since the reward is **K units**, Bob wants to ensure that the **total money he invests is strictly less than K**.

Bob wants to bet on **as many horses as possible**.

**Task**

Find the **length of the maximum continuous sequence of horses** on which Bob can bet such that the **total betting price is less than K**.

---

Example

Input
N = 5
prices = [10, 20, 30, 40, 50]
K = 60


Output

2


Explanation


10 + 20 = 30 < 60
20 + 30 = 50 < 60
30 + 40 = 70 > 60


Maximum valid continuous sequence length is **2**.

---

### Core Concept

Both problems can be solved using the **Sliding Window (Two Pointer) technique**.
