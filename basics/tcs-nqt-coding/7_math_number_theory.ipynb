```python
"""
QUESTION: K-th Largest Factor

You are given two integers N and K.

A factor of a number N is any positive integer that divides N exactly
(i.e., N % x == 0).

Your task is to find the K-th largest factor of N.

If N has fewer than K factors, return 1.

----------------------------------------

Input Format:
- A single line containing two integers N and K

----------------------------------------

Output Format:
- Print a single integer representing the K-th largest factor of N

----------------------------------------

Constraints:
- 1 ≤ N ≤ 10^9
- 1 ≤ K ≤ 10^5

----------------------------------------

Example 1:

Input:
N = 12, K = 2

Output:
6

Explanation:
Factors of 12: 1, 2, 3, 4, 6, 12  
Sorted (descending): 12, 6, 4, 3, 2, 1  
2nd largest = 6

----------------------------------------

Example 2:

Input:
N = 10, K = 4

Output:
1

Explanation:
Factors of 10: 1, 2, 5, 10  
Sorted (descending): 10, 5, 2, 1  
4th largest = 1

----------------------------------------

Example 3:

Input:
N = 7, K = 3

Output:
1

Explanation:
Factors of 7: 1, 7  
Only 2 factors exist, which is less than K = 3  
So, return 1

----------------------------------------

Note:
- Consider only positive factors.
- Order factors in descending order to determine the K-th largest.
"""
```

```python
def kthLargestFactor(N, k):

    factors = []

    for i in range(1, int(N**0.5) + 1):
        if N % i == 0:
            factors.append(i)

            if N//i != i:
                factors.append(N//i)

    factors.sort(reverse=True)

    if k>len(factors):
        return 1
    else:
        return factors[k-1]
print(kthLargestFactor(36,2))
```

```python
"""
QUESTION: Square-Free Numbers

A number is called square-free if it is not divisible by any perfect square
greater than 1 (i.e., no prime factor appears more than once in its factorization).

You are given an integer N.

Your task is to find how many divisors of N are square-free.

----------------------------------------

Input Format:
- A single integer N

----------------------------------------

Output Format:
- Print a single integer representing the count of square-free divisors of N

----------------------------------------

Constraints:
- 1 ≤ N ≤ 10^9

----------------------------------------

Example 1:

Input:
N = 12

Output:
4

Explanation:
Divisors of 12: 1, 2, 3, 4, 6, 12

Square-free divisors:
1 (✔)
2 (✔)
3 (✔)
4 (✖, divisible by 2²)
6 (✔)
12 (✖, divisible by 2²)

Count = 4

----------------------------------------

Example 2:

Input:
N = 18

Output:
4

Explanation:
Divisors of 18: 1, 2, 3, 6, 9, 18

Square-free divisors:
1 (✔)
2 (✔)
3 (✔)
6 (✔)
9 (✖, divisible by 3²)
18 (✖, divisible by 3²)

Count = 4

----------------------------------------

Example 3:

Input:
N = 1

Output:
1

Explanation:
Only divisor is 1, which is square-free.

----------------------------------------

Note:
- A perfect square is a number like 4, 9, 16, 25, etc.
- A divisor is square-free if none of its prime factors appear more than once.
"""
```

```python
def allFactors(N):
    factors = []

    for i in range(1, int(N**0.5)+1):

        if N%i == 0:
            factors.append(i)

            if N//i != i:
                factors.append(N//i)
    return factors

def isSquareFree(num):
    for i in range(2, int(num**0.5)+1):
        if num % (i*i) == 0:
            return False
    return True


def squareFreeCount(N):
    square_free_count = 0

    factors = allFactors(N)

    for i in range(len(factors)):
        if isSquareFree(factors[i]):
            square_free_count += 1
    
    return square_free_count

print(squareFreeCount(18))
```
