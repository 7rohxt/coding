```python
"""
QUESTION: Balance '*' and '#' in a String

You are given a string S consisting only of two characters:
'*' (asterisk) and '#' (hash).

Your task is to determine the balance between these two characters.

----------------------------------------

Definition of Balance:
- Return a positive integer if the number of '*' is greater than '#'
- Return a negative integer if the number of '#' is greater than '*'
- Return 0 if both counts are equal

The value returned should be:
    count('*') - count('#')

----------------------------------------

Input Format:
- A single string S

----------------------------------------

Output Format:
- Print an integer representing the balance

----------------------------------------

Constraints:
- 1 ≤ length of S ≤ 10^5
- S contains only '*' and '#'

----------------------------------------

Example 1:

Input:
S = "*#**"

Output:
2

Explanation:
Number of '*' = 3
Number of '#' = 1
Balance = 3 - 1 = 2

----------------------------------------

Example 2:

Input:
S = "##*#"

Output:
-2

Explanation:
Number of '*' = 1
Number of '#' = 3
Balance = 1 - 3 = -2

----------------------------------------

Example 3:

Input:
S = "*#*#"

Output:
0

Explanation:
Number of '*' = 2
Number of '#' = 2
Balance = 0

----------------------------------------

Note:
- This is a simple counting problem.
- No need for complex data structures.
"""
```

```python
from collections import Counter
def balanceSymbols(S):
    freq = Counter(S)
    res = freq['*'] - freq['#']
    return res

S = input().strip()
print(balanceSymbols(S))
```

```python
"""
QUESTION: String Transformation

You are given three strings: S1, S2, and S3.

Perform the following transformations:

1. In S1, replace all vowels (a, e, i, o, u — both lowercase and uppercase) with '%'.
2. In S2, replace all consonants (all alphabetic characters except vowels) with '#'.
3. In S3, convert all characters to uppercase.

After performing the transformations, concatenate the three resulting strings
in the same order: S1 + S2 + S3.

----------------------------------------

Input Format:
- First line: String S1
- Second line: String S2
- Third line: String S3

----------------------------------------

Output Format:
- Print the final concatenated string

----------------------------------------

Constraints:
- 1 ≤ length of each string ≤ 10^5
- Strings contain only alphabetic characters

----------------------------------------

Example:

Input:
hello
world
python

Output:
h%ll%####PYTHON

----------------------------------------

Note:
- Vowels are: a, e, i, o, u (case-insensitive)
- Consonants are all other alphabetic characters
"""
```

```python
def stringTrans(S1, S2, S3):
    vowels_set = {'a', 'e', 'i', 'o', 'u'}

    S1 = list(S1)
    S2 = list(S2)
    S3 = list(S3)

    for i in range(len(S1)):
        if S1[i].lower() in vowels_set:
            S1[i] = '%'

    for i in range(len(S2)):
        if S2[i].lower() not in vowels_set:
            S2[i] = '#'
    
    for i in range(len(S3)):
        S3[i] = S3[i].upper()

    res = S1+S2+S3

    return "".join(res)



S1 = input().strip()
S2 = input().strip()
S3 = input().strip()

print(stringTrans(S1, S2, S3))
```
