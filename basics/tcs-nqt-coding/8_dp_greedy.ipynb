```python
"""
QUESTION: House Robber Problem

You are given an array of integers representing the amount of money
in each house. There are N houses arranged in a line.

A thief wants to steal money but cannot rob two adjacent houses,
as it would alert the police.

Your task is to determine the maximum amount of money the thief can
steal without robbing two consecutive houses.

----------------------------------------

Input Format:
- First line: Integer N (number of houses)
- Second line: N space-separated integers representing money in each house

----------------------------------------

Output Format:
- Print a single integer representing the maximum amount that can be stolen

----------------------------------------

Constraints:
- 1 ≤ N ≤ 10^5
- 0 ≤ value of each house ≤ 10^4

----------------------------------------

Example 1:

Input:
N = 5
[2, 7, 9, 3, 1]

Output:
12

Explanation:
Rob house 1 (2), house 3 (9), and house 5 (1)
Total = 2 + 9 + 1 = 12

----------------------------------------

Example 2:

Input:
N = 4
[1, 2, 3, 1]

Output:
4

Explanation:
Rob house 1 and house 3 → 1 + 3 = 4

----------------------------------------

Example 3:

Input:
N = 3
[2, 1, 1]

Output:
3

Explanation:
Rob house 1 and house 3 → 2 + 1 = 3

----------------------------------------

Note:
- You cannot pick two adjacent indices.
- This is a classic dynamic programming problem.
"""
```

```python
def houseRobber(N, nums):

    if N == 0:
        return 0
    if N==1:
        return nums[1]
        
    dp = [0] * N
    dp[0] = nums[0]
    dp[1] = max(nums[0], nums[1]) 

    for i in range(2, N):
        dp[i] = max(dp[i-2] + nums[i], dp[i-1])

    return dp[N-1]

N = int(input().strip())
nums = list(map(int, input().split()))

print(houseRobber(N, nums))
```

```python
"""
QUESTION: Two Wheeler and Four Wheeler Production

A factory produces two types of vehicles:
- Two-wheelers (each has 2 wheels)
- Four-wheelers (each has 4 wheels)

You are given:
- V → total number of vehicles
- W → total number of wheels

Your task is to determine how many two-wheelers and four-wheelers
were produced.

----------------------------------------

Input Format:
- A single line containing two integers V and W

----------------------------------------

Output Format:
- Print two integers:
    number_of_two_wheelers number_of_four_wheelers

----------------------------------------

Constraints:
- 0 ≤ V ≤ 10^6
- 0 ≤ W ≤ 10^7

----------------------------------------

Example 1:

Input:
V = 200, W = 540

Output:
130 70

Explanation:
130 two-wheelers → 130 × 2 = 260 wheels  
70 four-wheelers → 70 × 4 = 280 wheels  
Total = 260 + 280 = 540

----------------------------------------

Example 2:

Input:
V = 10, W = 40

Output:
0 10

Explanation:
All vehicles are four-wheelers.

"""
```

```python
def findVehicles(V, W):
    # x + y = V
    # 2x + 4y = W
    y = (W-(2*V))//2
    x = V - y
    return [x,y]
V, W = map(int, input().split())
print(*findVehicles(V,W))
```
